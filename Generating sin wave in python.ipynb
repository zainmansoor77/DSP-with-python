```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0,10,1000)
x = np.sin(2*np.pi*.25*t)
plt.plot(t,x)
plt.xlabel("Time in sec")
plt.ylabel("Amplitude")
plt.grid()
plt.legend("0.25 Hz")
plt.xlim(0,10)
plt.ylim(-1,1)
plt.axis("on")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

t = np.linspace(0,10,1000)
x = signal.sine(2*np.pi*.25*t)
plt.plot(t,x)
plt.xlabel("Time in sec")
plt.ylabel("Amplitude")
plt.grid()
plt.legend("0.25 Hz")
plt.xlim(0,10)
plt.ylim(-1,1)
plt.axis("on")
plt.show()
```
