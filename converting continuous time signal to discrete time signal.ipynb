```python
from scipy import signal
import numpy as np
import matplotlib.pyplot as plt

ft = 100 # frequency of continuous time signal
fs = 1000 # Sampling frequency
t = np.arange(0,0.02,.000005)
xt = signal.square(2 * np.pi * ft * t)
sampling_freq = 0.1
n = np.arange(1,20,1)
xn = signal .square(2 * np.pi * (ft / fs) * n)  # (ft / fs) is frequency of discrete time signal
plt.subplot(2,1,1)
plt.plot(t,xt)
plt.xlabel("Time in sec")
plt.ylabel("Amplitude")
plt.title("Continuous time signal")
plt.legend("1 Hz")
plt.grid()
plt.show()
plt.subplot(2,1,2)
plt.stem(n,xn)
plt.xlabel("discrete time in sec")
plt.ylabel("Amplitude")
plt.title("Discrete time signal")
plt.grid()
plt.show()


```
